```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
c0 = 1.7
epsilon = 0.001
L = 500
k1 = np.logspace(-5, 4, 100)
```

```python
F11_constant = (c0**2 * epsilon**(2/3) * L**(8/3) * scipy.special.gamma(4/3))\
    / (8 * np.sqrt(np.pi) * scipy.special.gamma(17/6))

F11 = F11_constant / (1 + (L * k1)**2)**(4/3)
```

```python
plt.figure(figsize = (10, 6))
plt.semilogx(k1, k1 * F11)
plt.xlabel('k1')
plt.ylabel('k1 * F11')
plt.title('Analytical Spectrum')
plt.show()
```

```python
F_22_constant = (c0**2 * epsilon**(2/3) * L**(14/3) * scipy.special.gamma(7/3))\
    / (4 * np.sqrt(np.pi) * scipy.special.gamma(17/6))

F_22 = F_22_constant * k1**2 / ((1 + (L * k1)**2)**(7/3))
```
